```python
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
import pandas as pd
import time
import re
```

valid email checking function

```python
def is_valid_email(email):
    email = email.strip()
    pattern = r'^[\w\.-]+@[\w\.-]+$'
    return re.match(pattern, email) is not None
```

```python
sender_email = ""
password = ""
subject = "Application for internship as Software Developer"
with open("msgg.txt", "r") as file:
    message = file.read()
    
smtp_server = "smtp.gmail.com"
smtp_port = 587
```

```python
def sendmail():
    context = smtplib.SMTP(smtp_server, smtp_port)
    context.starttls()
    context.login(sender_email, password)

    df_read = pd.read_csv('maildone.csv')
    read_recipients = list(df_read['Email'])
    i = 0
    
    for recipient in recipients: 
        if(i > 50):
            break
            
        if recipient in read_recipients:
            # print(recipient, "Done Already")
            continue
        
        i+=1
        if is_valid_email(recipient) == False:
            read_recipients.append(recipient)
            print(i, recipient, "Not Done")
            continue
        
        msg = MIMEMultipart()
        msg['From'] = sender_email
        msg['Subject'] = subject
        msg.attach(MIMEText(message, 'html'))
        msg['To'] = recipient
        print(i, recipient)
          
        context.sendmail(sender_email, recipient, msg.as_string())
        read_recipients.append(recipient)
        
        df_read = pd.DataFrame(read_recipients, columns=['Email'])
        df_read.to_csv('maildone.csv')
    
    context.quit()
    df_read = pd.DataFrame(read_recipients, columns=['Email'])
    df_read.to_csv('maildone.csv')
    print("Total Mailed:", i)
    return i
```
